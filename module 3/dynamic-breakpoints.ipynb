# Dynamic breakpoints 

## Review

We discussed motivations for human-in-the-loop:

(1) `Approval` - We can interrupt our agent, surface state to a user, and allow the user to accept an action

(2) `Debugging` - We can rewind the graph to reproduce or avoid issues

(3) `Editing` - You can modify the state 

We covered breakpoints as a general way to stop the graph at specific steps, which enables use-cases like `Approval`

We also showed how to edit graph state, and introduce human feedback. 

## Goals

Breakpoints are set by the developer on a specific node during graph compilation. 

But, sometimes it is helpful to allow the graph **dynamically interrupt** itself!

This is an internal breakpoint, and [can be achieved using `NodeInterrupt`](https://langchain-ai.github.io/langgraph/how-tos/human_in_the_loop/dynamic_breakpoints/#run-the-graph-with-dynamic-interrupt).

This has a few specific benefits: 

(1) you can do it conditionally (from inside a node based on developer-defined logic).

(2) you can communicate to the user why its interrupted (by passing whatever you want to the `NodeInterrupt`).

Let's create a graph where a `NodeInterrupt` is thrown based upon length of the input.

````python
%%capture --no-stderr
%pip install --quiet -U langgraph langchain_openai langgraph_sdk
````

````python
from IPython.display import Image, display

from typing_extensions import TypedDict
from langgraph.checkpoint.memory import MemorySaver
from langgraph.errors import NodeInterrupt
from langgraph.graph import START, END, StateGraph

class State(TypedDict):
    input: str

def step_1(state: State) -> State:
    print("---Step 1---")
    return state

def step_2(state: State) -> State:
    # Let's optionally raise a NodeInterrupt if the length of the input is longer than 5 characters
    if len(state['input']) > 5:
        raise NodeInterrupt(f"Received input that is longer than 5 characters: {state['input']}")
    
    print("---Step 2---")
    return state

def step_3(state: State) -> State:
    print("---Step 3---")
    return state

builder = StateGraph(State)
builder.add_node("step_1", step_1)
builder.add_node("step_2", step_2)
builder.add_node("step_3", step_3)
builder.add_edge(START, "step_1")
builder.add_edge("step_1", "step_2")
builder.add_edge("step_2", "step_3")
builder.add_edge("step_3", END)

# Set up memory
memory = MemorySaver()

# Compile the graph with memory
graph = builder.compile(checkpointer=memory)

# View
display(Image(graph.get_graph().draw_mermaid_png()))
````

Let's run the graph with an input that's longer than 5 characters. 

````python
initial_input = {"input": "hello world"}
thread_config = {"configurable": {"thread_id": "1"}}

# Run the graph until the first interruption
for event in graph.stream(initial_input, thread_config, stream_mode="values"):
    print(event)
````

If we inspect the graph state at this point, we the node set to execute next (`step_2`).


````python
state = graph.get_state(thread_config)
print(state.next)
````

We can see that the `Interrupt` is logged to state.

````python
print(state.tasks)
````

We can try to resume the graph from the breakpoint. 

But, this just re-runs the same node! 

Unless state is changed we will be stuck here.

````python
for event in graph.stream(None, thread_config, stream_mode="values"):
    print(event)
````

````python
state = graph.get_state(thread_config)
print(state.next)
````

Now, we can update state.

````python
graph.update_state(
    thread_config,
    {"input": "hi"},
)
````

````python
for event in graph.stream(None, thread_config, stream_mode="values"):
    print(event)
````

# SELF-EXERCISE:

#### Developed a spacecraft‚Äôs AI system communicates with Mission Control. Dynamic breakpoints are triggered if the AI reports a critical anomaly.

````python
from IPython.display import Image, display
from typing_extensions import TypedDict
from langgraph.checkpoint.memory import MemorySaver
from langgraph.errors import NodeInterrupt
from langgraph.graph import START, END, StateGraph

class MissionState(TypedDict):
    ai_report: str
    control_response: str

def system_diagnostics(state: MissionState) -> MissionState:
    """Check spacecraft system status."""
    print("---Running System Diagnostics---")
    return {"control_response": f"Diagnostics received: '{state['ai_report']}'"}

def anomaly_detection(state: MissionState) -> MissionState:
    """Detect anomalies and decide whether to interrupt."""
    print("---Analyzing Anomalies---")
    report = state["ai_report"].lower()

    if "critical" in report or "oxygen" in report or "life support" in report:
        raise NodeInterrupt(f"Critical anomaly reported! Message: '{state['ai_report']}'")

    print("---No Critical Anomalies Detected---")
    return {"control_response": state["control_response"] + " | System nominal."}

def mission_confirmation(state: MissionState) -> MissionState:
    """Confirm continuation of mission."""
    print("---Confirming Mission Continuation---")
    return {"control_response": state["control_response"] + " | Proceeding with mission."}

builder = StateGraph(MissionState)
builder.add_node("system_diagnostics", system_diagnostics)
builder.add_node("anomaly_detection", anomaly_detection)
builder.add_node("mission_confirmation", mission_confirmation)

builder.add_edge(START, "system_diagnostics")
builder.add_edge("system_diagnostics", "anomaly_detection")
builder.add_edge("anomaly_detection", "mission_confirmation")
builder.add_edge("mission_confirmation", END)

memory = MemorySaver()

graph = builder.compile(checkpointer=memory)

display(Image(graph.get_graph().draw_mermaid_png()))
````

````python
initial_input_normal = {"ai_report": "All systems green. Crew stable."}
thread_config_normal = {"configurable": {"thread_id": "mission_normal_1"}}

for event in graph.stream(initial_input_normal, thread_config_normal, stream_mode="values"):
    print(event)
````

````python
initial_input_critical = {"ai_report": "Warning! Oxygen levels critical!"}
thread_config_critical = {"configurable": {"thread_id": "mission_interrupt_1"}}

for event in graph.stream(initial_input_critical, thread_config_critical, stream_mode="values"):
    print(event)
````

````python
state = graph.get_state(thread_config_critical)
print("Next step:", state.next)
print("Tasks log:", state.tasks)
````

````python
for event in graph.stream(None, thread_config_critical, stream_mode="values"):
    print(event)
````

````python
graph.update_state(
    thread_config_critical,
    {"ai_report": "Oxygen levels stabilized. Proceed with operations."},
)

for event in graph.stream(None, thread_config_critical, stream_mode="values"):
    print(event)
````
